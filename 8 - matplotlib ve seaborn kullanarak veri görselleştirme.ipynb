## Çizgi Grafik

Çizgi grafik, en basit ve en yaygın kullanılan veri görselleştirme tekniklerinden biridir. Bir çizgi grafiği, bilgileri düz çizgilerle bağlanan bir dizi veri noktası veya işaretçi olarak görüntüler. Daha iyi görsel netlik için çizgilerin ve işaretleyicilerin şeklini, boyutunu, rengini ve diğer estetik öğelerini özelleştirebiliriz.

Kanto adlı hayali bir ülkede altı yıl boyunca elma hasılatını (hektar başına ton) gösteren bir Python listesi.

````python
elmalarin_hasilati = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

Bir çizgi grafiği kullanarak elma hasılatının zaman içinde nasıl değiştiğini görselleştirebiliriz. Çizgi grafiği çizmek için `plt.plot` fonksiyonunu kullanabiliriz.

````python
plt.plot(elmalarin_hasilati)
````

`plt.plot` fonksiyonunun çağrılması, beklendiği gibi çizgi grafiğini çizer. Ayrıca, çıktıda gösterilen `[<matplotlib.lines.Line2D at 0x7ff70aa20760>]` çizilen grafiklerin bir listesini döndürür. Çıktıyı göstermekten kaçınmak ve yalnızca grafiği görüntülemek için hücredeki son ifadenin sonuna noktalı virgül (`;`) ekleyebiliriz.

````python
plt.plot(elmalarin_hasilati);
````

Daha bilgilendirici ve güzel hale getirmek için bu planı adım adım geliştirelim.

### X eksenini özelleştirme

Grafiğin X ekseni şu anda 0 ila 5 liste öğesi indekslerini gösterir. Verileri çizdiğimiz yılı gösterebilseydik, çizim daha bilgilendirici olurdu. Bunu iki argüman `plt.plot` ile yapabiliriz.

````python
yillar = [2010, 2011, 2012, 2013, 2014, 2015]
elmalarin_hasilati = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
````

````python
plt.plot(yillar, elmalarin_hasilati)
````

### Eksen Etiketleri

Her bir eksenin neyi temsil ettiğini göstermek için `plt.xlabel` ve `plt.ylabel` yöntemlerini kullanarak eksenlere etiketler ekleyebiliriz.

````python
plt.plot(yillar, elmalarin_hasilati)
plt.xlabel('Yıl')
plt.ylabel('Hasat (hektar başında ton)');
````

### Birden Çok Çizgi Çizme

Aynı grafikte birden çok çizgi çizmek için her çizgi için `plt.plot` fonksiyonunu bir kez çağırabiliriz. Kanto'daki elma ile portakalın hasılatını karşılaştıralım.

````python
yillar = range(2000, 2012)
elmalar = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
portakallar = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
````

````python
plt.plot(yillar, elmalar)
plt.plot(yillar, portakallar)
plt.xlabel('Yıl')
plt.ylabel('Hasılat (hektar başına ton)');
````

### Grafik Başlığı ve Açıklama/Gösterge

Birden çok satırı ayırt etmek için, `plt.legend` fonksiyonunu kullanarak grafiğin içine bir açıklama ekleyebiliriz. Ayrıca `plt.title` fonksiyonunu kullanarak grafik için bir başlık belirleyebiliriz.

````python
plt.plot(yillar, elmalar)
plt.plot(yillar, portakallar)

plt.xlabel('Yıl')
plt.ylabel('Hasılat (hektar başına ton)')

plt.title("Kanto'da Mahsul Verimi")
plt.legend(['Elmalar', 'Portakallar']);
````

### Çizgi İşaretleyiciler

Ayrıca, `plt.plot`'un `marker` argümanını kullanarak her satırdaki veri noktaları için işaretçileri gösterebiliriz. Matplotlib daire, çarpı, kare, elmas vb. gibi birçok farklı işaretleyici sağlar. İşaretleyici türlerinin tam listesini burada bulabiliriz: https://matplotlib.org/3.1.1/api/markers_api.html .

````python
plt.plot(yillar, elmalar, marker='o')
plt.plot(yillar, portakallar, marker='x')

plt.xlabel('Yıl')
plt.ylabel('Hasılat (hektar başına ton)')

plt.title("Kanto'da Mahsul Verimi")
plt.legend(['Elmalar', 'Portakallar']);
````

### Çizgileri ve İşaretleyicileri Şekillendirme

`plt.plot` fonksiyonu, çizgileri ve işaretçileri şekillendirme için birçok argümanı destekler:

- `color` or `c`: Çizginin rengini ayarlama. ([desteklenen renkler](https://matplotlib.org/3.1.0/gallery/color/named_colors.html))
- `linestyle` or `ls`: Düz veya kesikli çizgi arasında seçim yapma.
- `linewidth` or `lw`: Çizginin genişliğini ayarlama.
- `markersize` or `ms`: İşaretçilerin boyutunu ayarlama.
- `markeredgecolor` or `mec`: İşaretleyiciler için kenar rengini ayarlama.
- `markeredgewidth` or `mew`: İşaretçiler için kenar genişliğini ayarlama.
- `markerfacecolor` or `mfc`: İşaretçiler için dolgu rengini ayarlama.
- `alpha`: Çizginin opaklığı.

Daha fazla bilgi edinmek için `plt.plot` belgelerine bakabiliriz: [https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) .

````python
plt.plot(yillar, elmalar, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(yillar, portakallar, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Hasılat (hektar başına ton)')

plt.title("Kanto'da Mahsul Verimi")
plt.legend(['Elmalar', 'Portakallar']);
````

`fmt` argümanı, işaretçi şeklini, çizgi stilini ve çizgi rengini belirtmek için bir kısayol sağlar. `plt.plot` için üçüncü argüman olarak sağlanabilir.

```
fmt = '[işaretçi][çizgi][renk]'
```



````python
plt.plot(yillar, elmalar, 's-b')
plt.plot(yillar, portakallar, 'o--r')

plt.xlabel('Yıl')
plt.ylabel('Hasılat (hektar başına ton)')

plt.title("Kanto'da Mahsul Verimi")
plt.legend(['Elmalar', 'Portakallar']);
````

`fmt` içinde bir çizgi stili belirlemezseniz, yalnızca işaretçiler çizilir.

````python
plt.plot(yillar, portakallar, 'or')
plt.title("Portakalların Hasılatı (hektar başına ton)");
````

## Dağılım Grafiği

Bir dağılım grafiğinde, 2 değişkenin değerleri, 2 boyutlu bir ızgara üzerinde noktalar olarak çizilir. Ek olarak, noktaların boyutunu veya rengini belirlemek için üçüncü bir değişken de kullanabiliriz. Bir örnek deneyelim.

[İris çiçeği veri kümesi](https://en.wikipedia.org/wiki/Iris_flower_data_set) üç çiçek türünde çanak yapraklar ve taç yapraklar için örnek ölçümler sağlar. Iris veri seti Seaborn kütüphanesine dahildir ve bir Pandas veri çerçevesi olarak yüklenebilir.

````python
# Verileri bir Pandas veri çerçevesine yükleyelim.

cicekler_df = sns.load_dataset("iris")
````

````python
cicekler_df
````

````python
cicekler_df.species.unique()
````

Çanak yaprak uzunluğu ve çanak yaprak genişliği arasındaki ilişkiyi görselleştirmeye çalışalım. İlk içgüdümüz, `plt.plot` kullanarak bir çizgi grafiği oluşturmak olabilir.

````python
plt.plot(cicekler_df.sepal_length, cicekler_df.sepal_width);
````

Veri kümesi içinde iki özelliğin çok fazla kombinasyonu olduğu için çıktı çok bilgilendirici değildir. Aralarında basit bir ilişki yok gibi görünüyor.

Çanak yaprak uzunluğu ve çanak yaprak genişliğinin `seaborn` modülünden (`sns` olarak içe aktarılmış) `scatterplot` fonksiyonunu kullanarak nasıl değiştiğini görselleştirmek için bir dağılım grafiği kullanabiliriz.

````python
sns.scatterplot(x=cicekler_df.sepal_length, y=cicekler_df.sepal_width);
````

### Ton Ekleme

Yukarıdaki çizimdeki noktaların, bazı uç değerlerle nasıl farklı kümeler oluşturduğuna dikkat edelim. Çiçek türlerini `hue` olarak kullanarak noktaları renklendirebiliriz. Ayrıca `s` argümanını kullanarak noktaları büyütebiliriz.

````python
sns.scatterplot(x=cicekler_df.sepal_length, y=cicekler_df.sepal_width, hue=cicekler_df.species, s=100);
````

Tonlar eklemek, çizimi daha bilgilendirici hale getirir. Setosa çiçeklerinin çanak yaprak uzunluğunun daha küçük ama çanak yaprak genişliğinin daha yüksek olduğunu hemen söyleyebiliriz. Buna karşılık, Virginica çiçekleri için bunun tersi geçerlidir.

### Seaborn Şekillerini Özelleştirme

Seaborn, Matplotlib çizim fonksiyonlarını dahili olarak kullandığından, şekli değiştirmek için `plt.figure` ve `plt.title` gibi fonksiyonları kullanabiliriz.

````python
plt.figure(figsize=(12, 6))
plt.title('Çanak Yaprak Boyutları')

sns.scatterplot(x=cicekler_df.sepal_length, 
                y=cicekler_df.sepal_width, 
                hue=cicekler_df.species,
                s=100);
````

### Pandas Veri Çerçevelerini Kullanarak Çizim Yapma

Seaborn, Pandas veri çerçeveleri için yerleşik desteğe sahiptir. Her sütunu bir dizi olarak iletmek yerine, sütun adları sağlayabilir ve bir veri çerçevesi belirtmek için `data` bağımsız değişkenini kullanabiliriz.

````python
plt.title('Çanak Boyutları')
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species',
                s=100,
                data=cicekler_df);
````

## Histogram

Bir histogram, değerler aralığı boyunca bölmeler (aralık) oluşturarak ve her bölmedeki gözlem sayısını belirtmek için dikey çubuklar göstererek bir değişkenin dağılımını temsil eder.

Örneğin, çiçekler veri setindeki çanak yaprak genişlik değerlerinin dağılımını görselleştirelim. Bir histogram oluşturmak için `plt.hist` fonksiyonunu kullanabiliriz.

````python
# Verileri bir Pandas veri çerçevesine yükleyelim.
cicekler_df = sns.load_dataset("iris")
````

````python
cicekler_df.sepal_width
````

````python
plt.title("Çanak Yaprak Genişlik Dağılımı")
plt.hist(cicekler_df.sepal_width);
````

Çanak genişliklerinin 2.0 - 4.5 aralığında olduğunu ve yaklaşık 35 değerin 2.9 - 3.1 aralığında olduğunu hemen görebiliriz ki bu en kalabalık bin kutu görünüyor.

### Kutuların boyutunu ve sayısını kontrol etme

Bins argümanını kullanarak kutuların sayısını veya her birinin boyutunu kontrol edebiliriz.

````python
# Kutu sayısının belirtilmesi
plt.hist(cicekler_df.sepal_width, bins=5);
````

````python
import numpy as np

# Her bölmenin sınırlarını belirleme
plt.hist(cicekler_df.sepal_width, bins=np.arange(2, 5, 0.25));
````

````python
# Eşit olmayan boyutlarda kutular
plt.hist(cicekler_df.sepal_width, bins=[1, 3, 4, 4.5]);
````

### Çoklu Histogramlar

Çizgi grafiklere benzer şekilde, tek bir grafikte birden çok histogram çizebiliriz. Her bir histogramın opaklığını azaltabiliriz, böylece bir histogramın çubukları diğerlerini gizlemez.

Her çiçek türü için ayrı histogramlar çizelim.

````python
setosa_df = cicekler_df[cicekler_df.species == 'setosa']
versicolor_df = cicekler_df[cicekler_df.species == 'versicolor']
virginica_df = cicekler_df[cicekler_df.species == 'virginica']
````

````python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
plt.hist(virginica_df.sepal_width, alpha=0.4, bins=np.arange(2, 5, 0.25));
````

Ayrıca birden fazla histogramı üst üste yığabiliriz.

````python
plt.title('Sepal Genişlik Dağılımı')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

plt.legend(['Setosa', 'Versicolor', 'Virginica']);
````

## Grafik Çubuğu

Çubuk grafikler, çizgi grafiklere oldukça benzer, yani bir dizi değer gösterirler. Ancak, çizgilerle birbirine bağlanan noktalar yerine her değer için bir çubuk gösterilir. Çubuk grafiği çizmek için `plt.bar` fonksiyonunu kullanabiliriz.

````python
yillar = range(2000, 2006)
elmalar = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
portakallar = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
````

````python
plt.bar(yillar, portakallar);
````

Histogramlar gibi, çubukları üst üste yığabiliriz. Bunu yapmak için `plt.bar`'ın `bottom` argümanını kullanırız.

````python
plt.bar(yillar, elmalar)
plt.bar(yillar, portakallar, bottom=elmalar);
````

### Ortalamalı Çubuk Grafikler

Seaborn'da bulunan ve `tips` yani `ipuçları` adı verilen başka bir örnek veri kümesine bakalım. Veri seti, bir hafta boyunca bir restoranı ziyaret eden müşteriler için cinsiyet, günün saati, toplam fatura ve bahşiş miktarı hakkında bilgiler içerir.

````python
ipuclari_df = sns.load_dataset("tips");
````

````python
ipuclari_df
````

Ortalama fatura tutarının haftanın farklı günlerinde nasıl değiştiğini görselleştirmek için bir çubuk grafik çizmek isteyebiliriz. Bunu yapmanın bir yolu, günlük ortalamaları hesaplamak ve ardından `plt.bar`'ı kullanmaktır (bunu bir alıştırma olarak deneyelim).

Ancak, bu çok yaygın bir kullanım durumu olduğundan, Seaborn kütüphanesi, ortalamaları otomatik olarak hesaplayabilen bir `barplot` fonksiyonu sağlar.

````python
sns.barplot(x='day', y='total_bill', data=ipuclari_df);
````

## Heatmap

Renkleri kullanarak matris veya tablo gibi 2 boyutlu verileri görselleştirmek için bir ısı haritası (heatmap) kullanılır. Bunu anlamanın en iyi yolu bir örneğe bakmaktır. 12 yıl boyunca bir havaalanındaki aylık yolcu trafiğini görselleştirmek için Seaborn'dan `flights` adı verilen başka bir örnek veri kümesi kullanacağız.

````python
ucuslar_df = sns.load_dataset("flights").pivot("month", "year", "passengers")
````

````python
ucuslar_df
````

`ucuslar_df`, her ay için bir satır ve her yıl için bir sütun içeren bir matristir. Değerler, bir yılın belirli bir ayında havalimanını ziyaret eden yolcu sayısını (1000 olarak) göstermektedir. Havaalanındaki trafiği görselleştirmek için `sns.heatmap` fonksiyonunu kullanabiliriz.

````python
plt.title("Yolcuların Sayısı (1000'ler)")
sns.heatmap(ucuslar_df);
````

Daha parlak renkler, havaalanında daha fazla yolcu olduğunu gösterir. Grafiğe bakarak iki şey çıkarabiliriz:

- Herhangi bir yılda havaalanındaki yolcu trafiği, Temmuz ve Ağustos aylarında en yüksek olma eğilimindedir.
- Herhangi bir ayda havaalanındaki yolcu trafiği yıldan yıla artma eğilimindedir.

Ayrıca, renk paletini değiştirmek için `annot=True` belirterek ve `cmap` argümanını kullanarak her bloktaki gerçek değerleri görüntüleyebiliriz.

````python
plt.title("Yolcuların Sayısı (1000'ler)")
sns.heatmap(ucuslar_df, fmt="d", annot=True, cmap='Blues');
````

## Bir ızgarada birden çok grafik çizme

Matplotlib ve Seaborn ayrıca, çizim için bir dizi eksen döndüren `plt.subplots` kullanarak bir ızgarada birden çok grafiğin çizilmesini de destekler.

İşte bu eğitimde ele aldığımız farklı grafik türlerini gösteren tek bir ızgara.

````python
fig, axes = plt.subplots(2, 3, figsize=(16, 8))

# Çizim için eksenleri kullanalım.
axes[0,0].plot(yillar, elmalar, 's-b')
axes[0,0].plot(yillar, portakallar, 'o--r')
axes[0,0].set_xlabel('Yıllar')
axes[0,0].set_ylabel('Hasılat (hektar başına ton)')
axes[0,0].legend(['Elmalar', 'Portakallar']);
axes[0,0].set_title("Kanto'da Mahsul Hasılatı")


# Eksenleri Seaborn'a geçirelim.
axes[0,1].set_title('Çanak yaprak uzunluğu vs. Çanak yaprak genişliği')
sns.scatterplot(x=cicekler_df.sepal_length, 
                y=cicekler_df.sepal_width, 
                hue=cicekler_df.species, 
                s=100, 
                ax=axes[0,1]);

# Çizim için eksenleri kullanalım.
axes[0,2].set_title('Çanak Yaprak Genişlik Dağılımı')
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), 
         stacked=True);

axes[0,2].legend(['Setosa', 'Versicolor', 'Virginica']);

# Eksenleri Seaborn'a geçirelim.
axes[1,0].set_title('Restoran faturaları')
sns.barplot(x='day', y='total_bill', hue='sex', data=ipuclari_df, ax=axes[1,0]);

# Eksenleri Seaborn'a geçirelim.
axes[1,1].set_title('Uçuş Trafiği')
sns.heatmap(ucuslar_df, cmap='Blues', ax=axes[1,1]);

# Eksenleri kullanarak bir görüntü çizelim.
axes[1,2].set_title('Veri Bilimi Capsi')
axes[1,2].imshow(goruntu)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
````

Desteklenen fonksiyonların tam listesi için bu sayfaya bakalım:: https://matplotlib.org/3.3.1/api/axes_api.html#the-axes-class .
